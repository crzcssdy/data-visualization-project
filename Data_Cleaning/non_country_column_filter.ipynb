```python
import pandas as pd
import pycountry
import json

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('Resources/filtered_population_fertility_gdp_2014_2024.csv')
df.head()
```

```python
# Create a list of valid country names
valid_countries = [country.name for country in pycountry.countries]

# Filter rows based on country names
df = df[df['Country Name'].isin(valid_countries)]
df
```

```python
# Save the filtered DataFrame to a new CSV file
df.to_csv('Resources/fertility_gdp_2014_2024.csv', index=False)
# Save the filtered DataFrame to a JSON file
df.to_json('Resources/fertility_gdp_2014_2024.json', orient='records', indent=4)
```
